```python
data_Sun=pd.read_csv('SUNPHARMA.csv')### Sun Pharma
data_BTC=pd.read_csv('BTC.csv')### bitcoin
data_Lupin=pd.read_csv('LUPIN.csv') ### Lupin
data_Dog=pd.read_csv('DOGE.csv') ##DogeCoin
```

```python
data_Sun['Date']=pd.to_datetime(data_Sun['Date'])
data_BTC['Date']=pd.to_datetime(data_BTC['Date'])
data_Lupin['Date']=pd.to_datetime(data_Lupin['Date'])
data_Dog['Date']=pd.to_datetime(data_Dog['Date'])
```

```python
data_Sun['weekday']=data_Sun['Date'].dt.dayofweek
data_BTC['weekday']=data_BTC['Date'].dt.dayofweek
data_Lupin['weekday']=data_Lupin['Date'].dt.dayofweek
data_Dog['weekday']=data_Dog['Date'].dt.dayofweek
```

```python
data_Sun
```

```python
frame_Sun=data_Sun[(data_Sun['weekday']!=5)&(data_Sun['weekday']!=6)]
```

```python
frame_Sun.isna().sum()
```

```python
frame_BTC=data_BTC[(data_BTC['weekday']!=5)&(data_BTC['weekday']!=6)]
frame_BTC.isna().sum()
```

```python
frame_BTC[pd.isnull(frame_BTC).any(axis=1)]
```

```python
###filling all the nana with the moving averages of respective columns
frame_BTC['Open'][713]=frame_BTC.Open.rolling(window=2,min_periods=1).mean()[713]
frame_BTC['High'][713]=frame_BTC.High.rolling(window=2,min_periods=1).mean()[713]
frame_BTC['Low'][713]=frame_BTC.Low.rolling(window=2,min_periods=1).mean()[713]
frame_BTC['Close'][713]=frame_BTC.Close.rolling(window=2,min_periods=1).mean()[713]
frame_BTC['Volume'][713]=frame_BTC.Volume.rolling(window=2,min_periods=1).mean()[713]
frame_BTC['Adj Close'][713]=frame_BTC['Adj Close'].rolling(window=2,min_periods=1).mean()[713]
```

```python
frame_Lupin=data_Lupin[(data_BTC['weekday']!=5)&(data_BTC['weekday']!=6)]
frame_Lupin.isna().sum()
```

```python
frame_Lupin[pd.isnull(frame_Lupin).any(axis=1)]
```

```python
###filling all the nana with the moving averages of respective columns
frame_Lupin['Open'][360]=frame_Lupin.Open.rolling(window=2,min_periods=1).mean()[360]
frame_Lupin['High'][360]=frame_Lupin.High.rolling(window=2,min_periods=1).mean()[360]
frame_Lupin['Low'][360]=frame_Lupin.Low.rolling(window=2,min_periods=1).mean()[360]
frame_Lupin['Close'][360]=frame_Lupin.Close.rolling(window=2,min_periods=1).mean()[360]
frame_Lupin['Volume'][360]=frame_Lupin.Volume.rolling(window=2,min_periods=1).mean()[360]
frame_Lupin['Adj Close'][360]=frame_Lupin['Adj Close'].rolling(window=2,min_periods=1).mean()[360]
```

```python
frame_Dog=data_Dog[(data_BTC['weekday']!=5)&(data_BTC['weekday']!=6)]
frame_Dog.isna().sum()
```

```python
w=frame_Dog[pd.isnull(frame_Dog).any(axis=1)]
NA=w.index
w
```

```python
for i in w.index:
    frame_Dog['Open'][i]=frame_Dog.Open.rolling(window=2,min_periods=1).mean()[i]
    frame_Dog['High'][i]=frame_Dog.High.rolling(window=2,min_periods=1).mean()[i]
    frame_Dog['Low'][i]=frame_Dog.Low.rolling(window=2,min_periods=1).mean()[i]
    frame_Dog['Close'][i]=frame_Dog.Close.rolling(window=2,min_periods=1).mean()[i]
    frame_Dog['Volume'][i]=frame_Dog.Volume.rolling(window=2,min_periods=1).mean()[i]
    frame_Dog['Adj Close'][i]=frame_Dog['Adj Close'].rolling(window=2,min_periods=1).mean()[i]
```

```python
frame_Dog
```

```python
frame1=frame_Sun.iloc[:,1:7]
frame2=frame_BTC.iloc[:,1:7]
frame3=frame_Lupin.iloc[:,1:7]
frame4=frame_Dog.iloc[:,1:7]
```

```python
data_Sun['Close']
```

```python
k=frame_Sun['Close'].shift().fillna(data_Sun['Close'][0])
log_return_Sun=pd.DataFrame(np.log(data_Sun['Close']/k))
log_return_Sun
```

```python
l=frame_BTC['Close'].shift().fillna(data_Sun['Close'][0])
log_return_BTC=pd.DataFrame(np.log(frame_BTC['Close']/l))
log_return_BTC.reset_index(drop=True)
```

```python
m=frame_Lupin['Close'].shift().fillna(data_Sun['Close'][0])
log_return_Lupin=pd.DataFrame(np.log(frame_Lupin['Close']/m))
log_return_Lupin.reset_index(drop=True)
```

```python
n=frame_Dog['Close'].shift().fillna(data_Sun['Close'][0])
log_return_Dog=pd.DataFrame(np.log(frame_Dog['Close']/n))
log_return_Dog.reset_index(drop=True)
```

```python
scaler=StandardScaler()
scaled_return_Sun=scaler.fit_transform(log_return_Sun)
scaled_return_BTC=scaler.fit_transform(log_return_BTC)
scaled_return_Lupin=scaler.fit_transform(log_return_Lupin)
scaled_return_Dog=scaler.fit_transform(log_return_Dog)
```

```python
scaled_return_Sun=pd.DataFrame(scaled_return_Sun, columns=log_return_Sun.columns)
scaled_return_BTC=pd.DataFrame(scaled_return_BTC, columns=log_return_BTC.columns)
scaled_return_Lupin=pd.DataFrame(scaled_return_Lupin, columns=log_return_Lupin.columns)
scaled_return_Dog=pd.DataFrame(scaled_return_Dog, columns=log_return_Dog.columns)
```

```python
plt.title("log return Sun Pharma")
plt.plot(scaled_return_Sun)
```

```python
plt.title("log return Lupin")
plt.plot(scaled_return_Lupin)
```

```python
plt.title("log return Dog")
plt.plot(scaled_return_Dog)
```

```python
Sun_Scaled=scaler.fit_transform(frame1)
BTC_Scaled=scaler.fit_transform(frame2)
Lupin_Scaled=scaler.fit_transform(frame3)
Dog_Scaled=scaler.fit_transform(frame4)
```

```python
Sun_Scaled=pd.DataFrame(Sun_Scaled, columns=frame1.columns)
BTC_Scaled=pd.DataFrame(BTC_Scaled, columns=frame2.columns)
Lupin_Scaled=pd.DataFrame(Lupin_Scaled, columns=frame3.columns)
Dog_Scaled=pd.DataFrame(Dog_Scaled, columns=frame4.columns)
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in Sun_Scaled:
    plt.title('Sun_Scaled')
    ax = plt.subplot(4,4,plotnumber)
    plt.plot(Sun_Scaled[column])
    plt.xlabel(column,fontsize=20)   
    plotnumber+=1
```
